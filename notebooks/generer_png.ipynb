```python
from tensorflow.keras.datasets import mnist
from PIL import Image
import numpy as np

# Cargar dataset MNIST
(x_train, y_train), _ = mnist.load_data()

# Tomar un ejemplo cualquiera (aquí índice 0)
img = x_train[0]   # 28x28
label = y_train[0]

# Guardar como PNG
Image.fromarray(img).save(f"mnist_sample_{label}.png")

print("Listo. Se guardó la imagen:", f"mnist_sample_{label}.png")
```
